```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('matches.csv')
df
```

```python
df= df[['toss_winner', 'toss_decision', 'winner', 'venue']]
df
```

```python
df['win'] = [1 if df['toss_winner'][i]==df['winner'][i] else 0 for i in range(df.shape[0])]
df.drop(columns=['toss_winner', 'winner'], inplace=True)
df
```

```python
toss_decision_dummies = pd.get_dummies(df.toss_decision)
toss_decision_dummies = toss_decision_dummies.iloc[:, 1:]
toss_decision_dummies
```

```python
venue_dummies = pd.get_dummies(df.venue, prefix='Venue')
venue_dummies = venue_dummies.iloc[:, 1:]
venue_dummies
```

```python
df = pd.concat([toss_decision_dummies, venue_dummies, df], axis=1)
df.drop(columns=['toss_decision', 'venue'], inplace=True)
df
```

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
# from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05)
```

```python
# from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', n_jobs=-1)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
# from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
# from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
# from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
