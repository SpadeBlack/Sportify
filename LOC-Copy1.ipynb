```python
import requests

def bowl_type(player):
    url_player = 'https://cricapi.com/api/playerFinder?apikey=dafmauXJ13ZiapXIFOTbR6PLb1w1&name={}'
    outcome = requests.get(url_player.format(player.split()[-1])).json()

    url_stats = 'https://cricapi.com/api/playerStats?apikey=dafmauXJ13ZiapXIFOTbR6PLb1w1&pid={}'
    result = requests.get(url_stats.format(outcome['data'][0]['pid'])).json()

    return result['bowlingStyle']
```

```python
temp = []
for i in bowler['bowler']:
    try:
        temp.append(bowl_type(i))
    except:
        temp.append('')
    
bowler['type'] = temp

bowler
```
