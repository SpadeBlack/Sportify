## SRH bowlers analyse

## SRH powerplay bowlers

```python
powerplay_bowlers = balls[(balls.over>=1) & (balls.over<=6) & (balls.player_dismissed.notnull()) & (balls.bowling_team == 'Sunrisers Hyderabad') & (balls.is_super_over==0) & (balls.dismissal_kind!='run out')]
```

```python
powerplay_bowlers.head()
```

```python
top_powerplay_bowlers = powerplay_bowlers.groupby(['bowler']).count()
#bowlers['bowler'].sum()
```

```python
top_powerplay_bowlers
```

```python
top_powerplay_bowlers.columns
```

```python
top_powerplay_bowlers = top_powerplay_bowlers.drop(['inning', 'batting_team', 'bowling_team', 'over', 'ball',
       'batsman', 'non_striker', 'is_super_over', 'wide_runs', 'bye_runs',
       'legbye_runs', 'noball_runs', 'penalty_runs', 'batsman_runs',
       'extra_runs', 'total_runs', 'player_dismissed', 'dismissal_kind',
       'fielder'],axis=1)
```

```python
top_powerplay_bowlers.columns=['wickets']
```

```python
top_powerplay_bowlers
```

## SRH Deathovers bowlers

```python
deathovers_bowlers = balls[(balls.over>=15) & (balls.over<=20) & (balls.player_dismissed.notnull()) & (balls.bowling_team == 'Sunrisers Hyderabad') & (balls.is_super_over==0) & (balls.dismissal_kind!='run out')]
```

```python
deathovers_bowlers.head()
```

```python
top_deathovers_bowlers = deathovers_bowlers.groupby(['bowler']).count()
```

```python
top_deathovers_bowlers
```

```python
top_deathovers_bowlers.columns
```

```python
top_deathovers_bowlers = top_deathovers_bowlers.drop(['inning', 'batting_team', 'bowling_team', 'over', 'ball',
       'batsman', 'non_striker', 'is_super_over', 'wide_runs', 'bye_runs',
       'legbye_runs', 'noball_runs', 'penalty_runs', 'batsman_runs',
       'extra_runs', 'total_runs', 'player_dismissed', 'dismissal_kind',
       'fielder'],axis=1)
```

```python
top_deathovers_bowlers.columns=['wickets']
```

```python
top_deathovers_bowlers
```
